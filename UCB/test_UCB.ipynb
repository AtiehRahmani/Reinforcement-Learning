# Comparing $\varepsilon$-greedy, Optimistic and UCB algorithms

```python
from optimistic import multi_armed_bandit_optimistic
from eps_greedy import multi_armed_bandit_eps_greedy
from UCB import multi_armed_bandit_ucb
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_bandits = 10
num_steps = 1000
epsilon = 0.1
num_cases = 500
Q__init=5
c=1
```

```python
avg_reward = np.zeros(num_steps)
print(f'running for epsilon={epsilon} Q=0') 
for i in range(num_cases):
    reward_history = multi_armed_bandit_eps_greedy(num_bandits, epsilon, num_steps)
    for j in range(num_steps):
        avg_reward[j] += reward_history[j]
         
for i in range(num_steps):
    avg_reward[i]/=num_cases
plt.plot(avg_reward,label=f'epsilon={epsilon} , Q=0')

##################optimistic#####################
avg_reward = np.zeros(num_steps)
print(f'running for epsilon={epsilon} Q={Q__init}') 
for i in range(num_cases):
    reward_history = multi_armed_bandit_optimistic(num_bandits, epsilon, num_steps,Q__init)
    for j in range(num_steps):
        avg_reward[j] += reward_history[j]
         
for i in range(num_steps):
    avg_reward[i]/=num_cases
plt.plot(avg_reward,label=f'epsilon={epsilon} , Q=0')

##################ucb#####################
avg_reward = np.zeros(num_steps)
print(f'running for UCB') 
for i in range(num_cases):
    reward_history = multi_armed_bandit_ucb(num_bandits, num_steps,c)
    for j in range(num_steps):
        avg_reward[j] += reward_history[j]
         
for i in range(num_steps):
    avg_reward[i]/=num_cases
plt.plot(avg_reward,label=f'ucb')
plt.legend()
plt.xlabel("Step")
plt.ylabel("Average Rewards")
plt.show()
```

### The result shows that UCB is better than the two methods. Because UCB explores more to systematically reduce uncertainty, but its exploration reduces over time. Thus, we can say that UCB obtains a greater reward on average than other algorithms such as Epsilon-greedy, Optimistic Initial Values, etc.
