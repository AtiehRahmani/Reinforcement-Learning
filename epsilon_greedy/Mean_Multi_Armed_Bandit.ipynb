# epsilon-greedy Multi-Armed Bandit 
First, it runs for M cases (num_cases==500), where each case consists of N rewards (num_steps==1000). Then plot the average rewards of steps at each case across all cases: $\bar{1},\ \bar{2},\ \bar{3},\ \bar{4},...,\ \bar{N}$

```python
from eps_greedy import multi_armed_bandit_eps_greedy
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_bandits = 10
num_steps = 1000
epsilon = 0.1
avg_reward = np.zeros(num_steps)
num_cases = 500

for i in range(num_cases):
    reward_history = multi_armed_bandit_eps_greedy(num_bandits, epsilon, num_steps)
    for j in range(num_steps):
        avg_reward[j] += reward_history[j]
         
for i in range(num_steps):
    avg_reward[i]/=num_cases
```

```python

plt.plot(avg_reward)
plt.xlabel("Step")
plt.ylabel("Average Rewards")
plt.show()
```

## Change of epsilon in the above code:
The parameter epsilon ($\varepsilon$) determines the balance between exploration and exploitation. Epsilon represents the probability of choosing a random action (exploration), while 1-$\varepsilon$ represents the probability of choosing the action with the highest estimated value (exploitation).

```python
num_bandits = 10
num_steps = 1000
epsilons = [0,0.1,0.9]
avg_reward = np.zeros(num_steps)
num_cases = 500

for epsilon in epsilons:  
    for i in range(num_cases):
        reward_history = multi_armed_bandit_eps_greedy(num_bandits, epsilon, num_steps)
        for j in range(num_steps):
            avg_reward[j] += reward_history[j]
         
    for i in range(num_steps):
        avg_reward[i]/=num_cases
    plt.plot(avg_reward,label=f'epsilon={epsilon}')
plt.legend()
plt.xlabel("Step")
plt.ylabel("Average Rewards")
plt.show()
```
